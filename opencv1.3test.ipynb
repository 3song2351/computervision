### Making Borders for Images (Padding)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
BLUE = [33,0,0]
img1 = cv2.imread('logo.jpg')

replicate = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REPLICATE)
reflect = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT)
reflect101 = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_REFLECT_101)
wrap = cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_WRAP)
constant= cv2.copyMakeBorder(img1,10,10,10,10,cv2.BORDER_CONSTANT,value=BLUE)

plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.show()
```

### 1.3.2 Arithmetic Operations on Images

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('logo222.jpg')
img2 = cv2.imread('logo22.jpg')

dst = cv2.addWeighted(img1,0.7,img2,0.3,0)

cv2.imshow('dst',dst)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### 1.3.3 Performance Measurement and Improvement Techniques

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def nothing(x):
    pass

cv2.namedWindow('image1')
cv2.createTrackbar('medianBlur','image1',5,49,nothing)
img1 = cv2.imread('mountain.jpg')

i=5
while(1):
    i=cv2.getTrackbarPos('medianBlur','image1')
    i=int(i/2)*2+1;
    dst = cv2.medianBlur(img1,i)
    cv2.imshow('image1',dst)
    k=cv2.waitKey (1)& 0xFF
    if (k == 27) :
        break
        

cv2.destroyAllWindows()
```
